```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('cleaned-weather-data.csv')

dataset.head()
```

```python
month=dataset.DATE.str.split('-',expand=True)[1]
day=dataset.DATE.str.split('-',expand=True)[2]
dataset['month']=month
dataset['day']=day
dataset
```

```python
dataset.columns
```

```python
dataset_new=dataset.loc[:,['day','month','AVE_GSW']]
dataset_new
```

```python
s=np.arange(1,25)

s = pd.Series(s)
```

```python
df = pd.concat([s]*365, axis=0)
```

```python
df.index=dataset_new.index
df
```

```python
dataset_new['time']=df
solar_data_nys = dataset_new.rename(columns = {"AVE_GSW": "solar"})
```

```python
dataset_new.loc[1:50,"AVE_GSW"]
```
