```python
def wind_turbine_AWT26(x):
        # Importing the libraries
    import numpy as np
    import matplotlib.pyplot as plt
    import pandas as pd
    from sklearn.linear_model import LinearRegression
    
    # Considering the increase in wind speed with height
    # Shear ratio has been assumed
    height_of_turbine = 46 
    alpha = 0.14
    height_of_reading = 1 # in meters
    x_hub = x * ( height_of_turbine/height_of_reading)**alpha

    # Importing the dataset
    dataset = pd.read_csv('wind-solar-functions/power-curves.csv')
    X = dataset.iloc[:, 0].values.reshape(-1,1)
    y = dataset.iloc[:, 1].values


    # Training the Polynomial Regression model on the whole dataset
    from sklearn.preprocessing import PolynomialFeatures
    poly_reg = PolynomialFeatures(degree = 10)
    X_poly = poly_reg.fit_transform(X)
    lin_reg_2 = LinearRegression()
    lin_reg_2.fit(X_poly, y)


    # Predicting a new result with Polynomial Regression
    if x<=20.5:
        y = lin_reg_2.predict(poly_reg.fit_transform([[x_hub]]))
        y = float(y)
    
    else:
        y = lin_reg_2.predict(poly_reg.fit_transform([[20.5]]))
        y = float(y)
        
    return(y)
```

```python
solar = pd.read_csv("solar_data_nys.csv")
```

```python
# Removing the last row to make the dimensions the same:
solar.drop(df.tail(0).index,inplace=True)
```

```python
# Adding the solar column in data frame
df['solar'] = solar['solar']
```

```python
# Loading the solar functions:

def SolarConverter_AllMax_300(solar_radiation, temperature):
    NOCT = 44 # Nominal operating cell temperature
    power_max_at_1000 = 300 # power produced when cell temperature is 20 C
    temp_coefficient = 0.39 # percent per degree celcius
    cell_temperature = temperature + ((NOCT - 20)/800 )*solar_radiation # Cell temperature
    percent_power_reduction = 0.39*(cell_temperature - 20) # Percent power reduced from 20 --> cell temp
    power_at_NOCT_1000_radiation = (1 - percent_power_reduction/100)*power_max_at_1000 # Power produced when solar radiation is 1000 w/m2
    actual_power = power_at_NOCT_1000_radiation*solar_radiation/1000
    return actual_power


def SolarConverter_TrinaSolar_335(solar_radiation, temperature):
    NOCT = 44 # Nominal operating cell temperature
    power_max_at_1000 = 335 # power produced when cell temperature is 20 C
    temp_coefficient = 0.41 # percent per degree celcius
    cell_temperature = temperature + ((NOCT - 20)/800 )*solar_radiation # Cell temperature
    percent_power_reduction = 0.39*(cell_temperature - 20) # Percent power reduced from 20 --> cell temp
    power_at_NOCT_1000_radiation = (1 - percent_power_reduction/100)*power_max_at_1000 # Power produced when solar radiation is 1000 w/m2
    actual_power = power_at_NOCT_1000_radiation*solar_radiation/1000
    return actual_power

def SolarConverter_Mitsubishi(solar_radiation, temperature):
    NOCT = 45.7 # Nominal operating cell temperature
    power_max_at_1000 = 250 # power produced when cell temperature is 20 C
    temp_coefficient = 0.45 # percent per degree celcius
    cell_temperature = temperature + ((NOCT - 20)/800 )*solar_radiation # Cell temperature
    percent_power_reduction = 0.39*(cell_temperature - 20) # Percent power reduced from 20 --> cell temp
    power_at_NOCT_1000_radiation = (1 - percent_power_reduction/100)*power_max_at_1000 # Power produced when solar radiation is 1000 w/m2
    actual_power = power_at_NOCT_1000_radiation*solar_radiation/1000
    return actual_power

def SolarConverter_FirstSolar(solar_radiation, temperature):
    NOCT = 45 # Nominal operating cell temperature
    power_max_at_1000 = 435 # power produced when cell temperature is 20 C
    temp_coefficient = 0.32 # percent per degree celcius
    cell_temperature = temperature + ((NOCT - 20)/800 )*solar_radiation # Cell temperature
    percent_power_reduction = 0.39*(cell_temperature - 20) # Percent power reduced from 20 --> cell temp
    power_at_NOCT_1000_radiation = (1 - percent_power_reduction/100)*power_max_at_1000 # Power produced when solar radiation is 1000 w/m2
    actual_power = power_at_NOCT_1000_radiation*solar_radiation/1000
    return actual_power
```

```python
df['Solar_AllMax'] = df[['solar','hourly_temperature']].apply(lambda x: SolarConverter_AllMax_300(*x), axis=1)
df['Solar_TrinaSolar'] = df[['solar','hourly_temperature']].apply(lambda x: SolarConverter_TrinaSolar_335(*x), axis=1)
df['Solar_Mitsubishi'] = df[['solar','hourly_temperature']].apply(lambda x: SolarConverter_Mitsubishi(*x), axis=1)
df['Solar_FirstSolar'] = df[['solar','hourly_temperature']].apply(lambda x: SolarConverter_FirstSolar(*x), axis=1)
```
