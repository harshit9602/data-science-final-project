```python
import pandas as pd
import numpy as np
```

```python
df_combined = pd.read_csv('../final-applying-ga-to-all-cities/combined-cities-dataset.csv')
```

```python
wind_mw_cost = 1750000 # $1.75 mil per MW
wind_kw_cost = 6000 # $6000/kw for turbines <~ 100 kW

# for simplicity, we are only considering 5 kW systems:

solar_kw_cost = 10952 # $/kw for a 5 kW system
```

```python
awt26 = 275*wind_kw_cost     
awp = 200*wind_kw_cost
d2cf = 200*wind_kw_cost
an = 150*wind_kw_cost
aria = 20*wind_kw_cost
wwd = 1*wind_mw_cost

all_max=(5/300)/solar_kw_cost
trina=(5/335)/solar_kw_cost
mit=(5/250)/solar_kw_cost
first=(5/435)/solar_kw_cost
```
