```python
import pandas as pd
```

```python
df = pd.read_csv("../messy-electricity-consumption.csv")
df.__delitem__('Unnamed: 0')
df
```

```python
id_var = ["Year","Month","Day"]

t = pd.melt(df,id_var ,var_name="Hour", value_name="Consumption")

t = t.sort_values(by = ["Month","Day"])
```

```python
t = t.reset_index()
```

```python
index = (t['Month'] == 2) & (t['Day'] == 29)
index = -index
t = t[index]
t = t.reset_index()
```

```python
t
```

```python
t.to_csv('../cleaned-data-frames/target-electricity-consumption.csv')
```
