```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('ny-state-weather-data.csv')

dataset.head()
```

```python
time=dataset.DATE.str.split('T',expand=True)[1]
date=dataset.DATE.str.split('T',expand=True)[0]
dataset['TIME']=time
z=dataset.TIME.str.split(':',expand=True)[0]
dataset['TIME']=z
dataset['DATE']=date
month=dataset.DATE.str.split('-',expand=True)[0]
day=dataset.DATE.str.split('-',expand=True)[1]
dataset['MONTH']=month
dataset['DAY']=day
dataset.__delitem__('DATE')
dataset.head()
```

```python
different_types=[]
for i in dataset['NAME']:
  if i not in different_types:
    different_types.append(i)
print(different_types)
print(len(different_types))
#print(len(dataset['NAME']))
```

```python
# Chaging the 00 hours to 24

index = dataset.loc[:, "TIME"] == "00"

dataset.loc[index,"TIME"]= 24
```

```python
# Select the required file only 

dataset_master = dataset

dataset = dataset_master.loc[:, ["NAME","TIME","MONTH","DAY", 'HLY-TEMP-NORMAL','HLY-WIND-AVGSPD']]
```

```python
dataset = dataset.rename(columns={'HLY-TEMP-NORMAL': 'hourly_temperature', 
                                 'HLY-WIND-AVGSPD': "hourly_wind_speed",
                                 "NAME" : "name", "TIME": "time",
                                 "MONTH":"month", "DAY":"day"})
```

```python
#index = dataset.loc[:,'NAME'] == 'LA GUARDIA AIRPORT, NY US'
#df_la_guardia = dataset[index]
#df_la_guardia.head()
```

```python
df_nyc = dataset[dataset.loc[:,'name'] == 'LA GUARDIA AIRPORT, NY US']
df_buffalo = dataset[dataset.loc[:,'name'] == 'BUFFALO NIAGARA INTERNATIONAL, NY US']
df_albany = dataset[dataset.loc[:,'name'] == 'ALBANY INTERNATIONAL AIRPORT, NY US']
df_rochester = dataset[dataset.loc[:,'name'] == 'ROCHESTER GREATER INTERNATIONAL, NY US']
df_binghamton = dataset[dataset.loc[:,'name'] == 'BINGHAMTON GREATER AP, NY US']
df_islip = dataset[dataset.loc[:,'name'] == 'ISLIP LI MACARTHUR AIRPORT, NY US']
df_nj = dataset[dataset.loc[:,'name'] == 'ATLANTIC CITY INTERNATIONAL AIRPORT, NJ US']
df_syracuse = dataset[dataset.loc[:,'name'] == 'SYRACUSE HANCOCK INTERNATIONAL AIRPORT, NY US']
```

```python
# Saving the required datasets:

df_nyc.to_csv("df_nyc.csv")
df_buffalo.to_csv("df_buffalo.csv")
df_albany.to_csv("df_albany.csv")
df_rochester.to_csv("df_rochester.csv")
df_binghamton.to_csv("df_binghamton.csv")
```
